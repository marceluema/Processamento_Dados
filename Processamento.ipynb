# Processameto de dados

```python
#Immportar as bibilotecas pandas e numpy
#Usar o método pd.read_excel para ler o arquivo xlsx

import pandas as pd
import numpy as np
df = pd.read_excel ('Base_dados.xlsx')
```

```python
# A função list mostra uma lista com as variáves do data frame

list(df)
```

```python
# calcular a % passando normalize = True

round(df['Anote seu sexo:'].value_counts(ascending = False, normalize=True)*100,2)
```

```python
#Checar se os valores da variável "id" são únicos

df['Id'].is_unique
```

```python
# cruzamento unsado pd.crosstab, colocar "Columns" no normalize retona a % na coluna e index na linha

round(pd.crosstab(df['Anote seu sexo:'],
             df ['Pensando na simpatia do consultor, você diria que ele foi:'], normalize='index')*100,2)
```

```python
#describe traz as principais estatísticas

df['NPS'].describe()
```

```python
# round arredonda

round(df['NPS'].mean(),2)
```

```python
# contar a quantidade de linhas

df['Id'].count()
```

```python
#Gerar frequência por Estado

Estado = df['Estado/Região']
round(Estado.value_counts(ascending = False, normalize=True)*100,2)
```

```python
# para selecionar uma ou mais colunas em um dataframe usar [] dentro de []

df1= df[['Id','NPS']]
df1
```

```python
#Filtrar o dataFrane com base em um valor de uma coluna

filtro_1 = df['Estado/Região']=='PA'

df_filtro = df[filtro_1]
df_filtro.head()
```

```python
#Filtrar dataFrame com base em valores de duas colunas

filtro_colunas =  df[df['Estado/Região'] =='PA'] [df['NPS'] ==9]
filtro_colunas 
```

```python
#filtro com daus opções na mesma coluna

filtro_opcoes = df[(df['Estado/Região']=="PA") | (df['Estado/Região']=="SP")]
filtro_opcoes
```

```python
#Filtrar uma coluna com duas opções e outra coluna com uma opção

filtro_advanced = df[(df['Estado/Região']=="PA") | 
                       (df['Estado/Região']=="SP")] [df['NPS']==10]
filtro_advanced
```
